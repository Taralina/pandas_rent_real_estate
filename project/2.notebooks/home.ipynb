```python
counts = df.copy()
counts['Цена'] = df['Цена'].apply(lambda x: x.split(','))
```

```python
counts
```

```python
def price(x):
    communal = [] #коммуналка
    prepayment = [] #предоплата
    bail = [] #залог
    rent = [] #квартплата
    rental = [] #срок
    
    for i in x:

        
        if 'За месяц' in i:
            if '$' in i:
                numbers = (re.findall(r'\d+\.\d+|\d+', i))[0]
                rent = float(numbers[0]) * 97
            elif '€' in i:
                numbers = (re.findall(r'\d+\.\d+|\d+', i))[0]
                rent = float(numbers[0]) * 106
            else:
                numbers = re.findall(r'\d+\.\d+|\d+', i) 
                rent = float(numbers[0])


        
        elif 'Залог' in i or 'залог' in i:
            if '$' in i:
                numbers = (re.findall(r'\b\d+\b', i))[0]
                bail = int(numbers[0]) * 97
            elif '€' in i:
                numbers = (re.findall(r'\b\d+\b', i))[0]
                bail = int(numbers[0]) * 106
            else:
                numbers = re.findall(r'\b\d+\b', i) 
                bail = int(numbers[0])

        
        elif 'комм' in i or 'Комм' in i:
            if 'включены' in i:
                communal = 0.0
            else:
                numbers = (re.findall(r'\b\d+\b', i))[0]
                communal= float(numbers) if numbers else 0                
                
        
        elif 'Срок' in i or 'срок' in i:
            rental.append(i)

        
        elif 'Предоплата' in i:
            numbers = re.findall(r'\d+', i)
            prepayment = int(numbers[0])

         
    return pd.Series({
        'Квартплата': rent if rent else 0,
        'Залог': bail if bail else 0,
        'комм': communal if communal is not None else None,
        'Срок': ', '.join(rental) if rental else 0,
        'Предоплата': prepayment if prepayment else 0
    })

        

counts[['Квартплата', 'Залог', 'комм', 'Срок', 'Предоплата']] = counts['Цена'].apply(price)

counts['комм'] = counts['комм'].apply(lambda x: np.nan if x == [] else x)
counts['комм'].fillna(counts['комм'].mean(), inplace=True)
```
